```python
%matplotlib inline
#%%
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import math as math
import json

COLOR = 'white'
plt.rcParams['text.color'] = COLOR
plt.rcParams['axes.labelcolor'] = COLOR
plt.rcParams['xtick.color'] = COLOR
plt.rcParams['ytick.color'] = COLOR

def writeLUT(ys):
    intValues = list(map(round, ys))
    # intValues = np.round(np.array(ys), 0, list)
    json_string = json.dumps(intValues)
    print(json_string)
    
def visualize(formula, x_range):  
    x = np.array(x_range)  
    y = np.array(list(map(formula, x)))
    # y = formula(x)  # <- note now we're calling the function 'formula' with x
    plt.figure(figsize=(20,15))
    plt.plot(x, y)  
    plt.show()  
    writeLUT(y.tolist())
    
def quad(x):
    return x**2 / 255
def cubic(x):
    return x**3 / 255 / 255
def linear(x):
    return x * 0.4

def fade(a, b, crossfade):
    return (a * (1-crossfade)) + (b*(crossfade))


def transform(x: int):
    if x < 1: 
         return 0
    minimumValue = 87
    maximumValue = 255
    xPercentage = x/255
    crossfadeHalf = max(0, min(1, xPercentage*2 - 1))
    crossfadeQuad = x**2/255/255
    nonlinearValue = fade( linear(x), quad(x), 1)
    projectedValue = nonlinearValue * ((maximumValue - minimumValue) / 255)
    return projectedValue + minimumValue


visualize(transform, range(0,256))
# visualize(linear, range(0,255))
```
